```python
import pandas as pd
from telegram.ext import Updater, MessageHandler, Filters

# Read the Excel dataset
dataset = pd.read_excel('/content/firstaidmodel.xlsx', engine='openpyxl')

# Function to extract relevant information from the scenario
def extract_information(scenario, dataset):
    scenario = scenario.lower()
    do_list = []
    dont_list = []
    medication_list = []

    for index, row in dataset.iterrows():
        if row['Scenario'].lower() in scenario:
            do_list.append(row["Do's"])
            dont_list.append(row["Don'ts"])
            medication_list.append(row['Medications'])

    return do_list, dont_list, medication_list

# Function to handle messages from users
def handle_message(update, context):
    user_scenario = update.message.text
    do_list, dont_list, medication_list = extract_information(user_scenario, dataset)

    response = "Do's:\n"
    for do in do_list:
        response += "- " + do + "\n"

    response += "\nDon'ts:\n"
    for dont in dont_list:
        response += "- " + dont + "\n"

    response += "\nMedications:\n"
    for medication in medication_list:
        response += "- " + medication + "\n"

    context.bot.send_message(chat_id=update.effective_chat.id, text=response)

# Create the Telegram bot updater and dispatcher
updater = Updater("YOUR_BOT_TOKEN", use_context=True)
dispatcher = updater.dispatcher

# Register the message handler
message_handler = MessageHandler(Filters.text, handle_message)
dispatcher.add_handler(message_handler)

# Start the bot
updater.start_polling()
```
