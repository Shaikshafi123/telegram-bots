```python
import telebot
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Create a Telegram bot
bot = telebot.TeleBot("5897940772:AAHDQnR0oTmkoorgXwg3Ox9hp_6n_IyHlIA")

# Define a function to handle updates from the Telegram API
def handle_update(update):
  # Get the text of the message
  text = update.message.text

  # If the message is a command.
  if text.startswith("/"):
    # Handle the command.
    if text == "/brightness":
      # Get the brightness value from the message.
      brightness = int(text.split()[1])

      # Greet the user
      bot.send_message(update.message.chat.id, "Hello! How can I help you today?")

      # Get the image from the user.
      image = bot.get_file(update.message.photo[-1].file_id)
      image = open(image.file_path, 'rb')

      # Adjust the brightness of the image.
      image = cv2.addWeighted(image, 1, np.full(image.shape, brightness / 100, dtype=np.uint8), 0, 0)

      # Save the image.
      cv2.imwrite('Brightness.jpg', image)

      # Send the image to the user.
      bot.send_photo(update.message.chat.id, open('Brightness.jpg', 'rb'))
    elif text == "/contrast":
      # Get the contrast value from the message.
      contrast = int(text.split()[1])

      # Greet the user
      bot.send_message(update.message.chat.id, "Hello! How can I help you today?")

      # Get the image from the user.
      image = bot.get_file(update.message.photo[-1].file_id)
      image = open(image.file_path, 'rb')

      # Adjust the contrast of the image.
      image = cv2.convertScaleAbs(image, alpha=contrast, beta=0)

      # Save the image.
      cv2.imwrite('Contrast.jpg', image)

      # Send the image to the user.
      bot.send_photo(update.message.chat.id, open('Contrast.jpg', 'rb'))


```
