```python
import telebot
from cryptography.fernet import Fernet

# Create a bot object
bot = telebot.TeleBot("6063716117:AAE1fs6sIqucaX2upLvnPqW2ntXokbPh9oo")

# Read the key from the file
with open("key.txt", "rb") as f:
    key = f.read()

# Create a function to encrypt text
def encrypt(text):
    f = Fernet(key)
    return f.encrypt(text.encode()).decode()

# Create a function to decrypt text
def decrypt(text):
    f = Fernet(key)
    return f.decrypt(text.encode()).decode()

# Create a handler for the /encrypt command
@bot.message_handler(commands=["encrypt"])
def encrypt_message(message):
    # Get the text from the message
    text = message.text.split(' ', 1)[1]

    # Encrypt the text
    encrypted_text = encrypt(text)

    # Send the encrypted text back to the user
    bot.send_message(message.chat.id, encrypted_text)

# Create a handler for the /decrypt command
@bot.message_handler(commands=["decrypt"])
def decrypt_message(message):
    # Get the text from the message
    text = message.text.split(' ', 1)[1]

    # Decrypt the text
    decrypted_text = decrypt(text)

    # Send the decrypted text back to the user
    bot.send_message(message.chat.id, decrypted_text)

# Start the bot
bot.polling()
```
